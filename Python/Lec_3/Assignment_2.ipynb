We all know about **Mandelbrot Set** right?,

<img src="mandel.jpg" alt="drawing" style="width:400px;"/>

To check any number in **Mandelbrot Set**, we take a number $c$. Then, 
we calculate $$f(z) = z^2 + c$$. Then we calculate $f(f(z))$ and so on and after some iteration the value becomes greater than or equals to 2, then it is not inside the set.

As an example, let's see if $c = 1$ is in the mandelbrot set.
1. We take $f(z) = z^2 + 1$.
2. Then, start the iteration with $z=0$, i.e., $f(0) = 0^2 + 1 = 1$.
3. Then, we take $z = 1$, i.e., previous result and calculate $f(z)$, i.e., $f(1) = 1^2 + 1 = 2$.
4. As the value becomes equal to $2$, hence $c = 1$ is not inside the mandelbrot set.

Write a code using for, if-else statement to check if any real or complex number inside the mandelbrot set for a max iteration number of 1000.
