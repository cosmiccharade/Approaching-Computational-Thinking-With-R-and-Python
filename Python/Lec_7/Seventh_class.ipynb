```python
from scipy.optimize import root
def pron(x):
    return np.tan(x) - x
```

```python
x_sols = [0]
for i in range(1,3):
    x_s = root(pron,(2*i-1)*np.pi/2 - 0.3).x[0]
    x_sols.append(x_s)
```

```python
plt.grid()
plt.plot(x,f(x),c="red",ls="--",label=r"$f(\theta) = \theta$")
plt.plot(x,g(x),c="blue",label=r"$g(\theta) = tan(\theta)$")
plt.scatter(x_sols,f(x_sols),color="black")
plt.ylim(-5,15)
plt.legend()
plt.show()
```

```python
x = np.linspace(0,15,30)
y_exp = np.sin(x) + 0.1*np.random.randn(len(x)) 
```

```python
plt.scatter(x,y_exp,color="blue",label="exp data.")
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x,y_exp,color="blue",marker="o",markeredgecolor="Blue",ls="--",label="exp data.")
plt.xlabel("Time(sec)")
plt.ylabel("Voltage")
plt.grid()
plt.legend()
plt.show()
```

```python
from scipy.optimize import curve_fit

def the_eq(x,a,b):
    return a*np.sin(x) + b
```

```python
para, error = curve_fit(the_eq,x,y_exp)
```

```python
para
```

```python
x_ax = np.linspace(0,15,1000)
plt.scatter(x,y_exp,color="blue",ls="--",label="exp data.")
plt.plot(x_ax,the_eq(x_ax,para[0],para[1]),ls="--",color="black",label="Theoret. express")
plt.xlabel("Time(sec)")
plt.ylabel("Voltage")
plt.grid()
plt.legend()
plt.savefig('gbhggbhy_plot.pdf')
plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,4))
ax.scatter(x,y_exp,color="blue",ls="--",label="exp data.")
ax.plot(x_ax,the_eq(x_ax,para[0],para[1]),ls="--",color="black",label="Theoret. express")
# plt.xlabel("Time(sec)")
# plt.ylabel("Voltage")
ax.set_xlabel("Time(sec)")
ax.set_ylabel("Voltage")
ax.grid()
ax.legend()
plt.savefig('gbhggbhy_plot.pdf')
plt.show()
```

```python
import matplotlib.pyplot as plt
import math as mp
pi_val = 3.1415926535897932
sqrt_3_3 = 3*mp.sqrt(3)

def leib_err(n):
    result = 0
    for i in range(n):
        result = result + 1/(4*i+1) - 1/(4*i+3)
    result = 4*result
    error = pi_val - result
    return result

def rou_err(n):
    result = 0
    for i in range(n):
        result = result + 1/(3*i+1) - 1/(3*i+2)
    result = result*sqrt_3_3
    error = pi_val - result
    return result
n_vals = [i for i in range(10_000+1)]
leibniz = [leib_err(i) for i in n_vals]
# print(leibniz)
rousan  =  [rou_err(i) for i in n_vals]
#
plt.plot(n_vals,leibniz,c='red',label='Leibniz Error')
plt.plot(n_vals,rousan,c='black',label='My formula Error')
plt.legend(loc='best',prop={'size':7})
plt.ylim(3.1375,3.1425)
plt.title("Error vs No. of terms for Leibniz and my $\\pi$ formula.")
plt.show()
```
